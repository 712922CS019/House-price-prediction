```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
data = fetch_california_housing()

# Convert to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df["Price"] = data.target

df  # sample code for dataset
```

```python
# no need of null value clearance and duplicate and other eda as we did already in the before sum

s=df.drop("Price",axis=1)
s
```

```python
o=df["Price"]
o
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor # Corrected import

# Ensure s and o are defined in the current scope
# Assuming 'df' is already defined from a previous cell execution
s=df.drop("Price",axis=1)
o=df["Price"]

train_x,test_x,train_y,test_y=train_test_split(s, o, test_size=0.2,random_state=42)
l=RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)
m=l.fit(train_x,train_y)
prediction=l.predict(test_x)
print("regressionfortrain",l.score(train_x,train_y))
print("regressionfortest",l.score(test_x,test_y))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(test_y,prediction)
plt.xlabel("test_y")
plt.ylabel("prediction")
plt.title("test vs prediction")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
mse=mean_squared_error(test_y,prediction)
rmse=np.sqrt(mse)
mae=mean_absolute_error(test_y,prediction)
print("mae",mae)
print("rmse",rmse)
print("mse",mse)
```

```python
from sklearn.model_selection import cross_val_score
cross=cross_val_score(l,s,o,cv=5,scoring="r2")
print("cross_score",cross)
print("cross_val_mean",cross.mean())
```

```python
# still the r2 values is low. lets go on with another algorithm
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score # Added import for r2_score

ParameterGrid={
    "n_estimators":[100,200,300],
    "learning_rate":[0.05,0.1],
    "max_depth":[2,3,4]
}
gbr=GradientBoostingRegressor(random_state=42)
grid=GridSearchCV(gbr,ParameterGrid,cv=5,scoring="r2",n_jobs=-1)
grid.fit(s,o)
print("best parameter",grid.best_params_)
print("best cv score",grid.best_score_)
best_model=grid.best_estimator_
pred=best_model.predict(test_x)
print("final_r2",r2_score(test_y,pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(test_y, pred)

plt.plot([test_y.min(), test_y.max()],
         [test_y.min(), test_y.max()],
         color='red')

plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Final Model Performance (GBR + GridSearch)")
plt.show()
```

```python
residuals = test_y - pred

plt.scatter(pred, residuals)
plt.axhline(y=0, color='red')

plt.title("Residual Plot")
plt.show()
```
